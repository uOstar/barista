```python
%matplotlib inline
from LogParser import LogParser
import matplotlib.pyplot as plt
plt.ion()
plt.style.use('ggplot')
lp = LogParser()
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (25, 10)

df = lp.getXs('1522127760.log.sample', 'Velocity')
df = df.set_index("Time")
df=df.astype(float)

ax1 = df.plot()
plt.xlabel("Time", fontsize="24")
plt.ylabel("X Acceleration (m/s^2)", fontsize="24")
```

```python
df = lp.getYs('1522127760.log.sample', 'Velocity')
df = df.set_index("Time")

%matplotlib inline
df=df.astype(float)
ax2 = df.plot()
ax2.autoscale()
```

```python
df = lp.getZs('1522127760.log.sample', 'Acceleration')
df = df.set_index("Time")

%matplotlib inline
df=df.astype(float)
ax3 = df.plot()
```
